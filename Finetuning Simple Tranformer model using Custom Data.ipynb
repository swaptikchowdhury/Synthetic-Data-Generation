```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")# specifying which model to use for tokenizing
```

```python
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True) # passing the text to tokenizer
```

```python
tokenized_datasets = dataset.map(tokenize_function, batched=True)
#https://huggingface.co/docs/datasets/about_map_batch
```

```python
small_train_dataset = tokenized_datasets["train"].shuffle(seed=42).select(range(1000))
small_eval_dataset = tokenized_datasets["test"].shuffle(seed=42).select(range(1000))
```

```python
# https://huggingface.co/docs/transformers/v4.25.1/en/main_classes/trainer#transformers
# Trainer class takes a pretrained model to retrain
```

```python
# Training with Pytorch
from transformers import AutoModelForSequenceClassification
model = AutoModelForSequenceClassification.from_pretrained("bert-base-cased", num_labels=5) # labels is a hyperparamater
```

```python
#Focusing on hyperparameters
#creates a TrainingArguments class which contains all the hyperparameters you can tune as well as flags for activating different training options.
#https://huggingface.co/docs/transformers/main_classes/trainer#transformers.TrainingArguments
from transformers import TrainingArguments
training_args = TrainingArguments(output_dir="test_trainer")
```

```python
!pip install evaluate
```

```python
# Evaluate. As Trainer class doesnt evaluate model during training, trainer has to attached with a model performance training function
import numpy as np
import evaluate
metric = evaluate.load("accuracy")
```

```python
#Call compute_metric to calculate the accuracy of your predictions.
#Before passing your predictions to compute, you need to convert the predictions to logits (remember all 🤗 Transformers models return logits):
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
```

```python
#To evaluation metrics during fine-tuning, specify the evaluation_strategy parameter in thetraining arguments to report the evaluation metric at the end of each epoch:
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(output_dir="test_trainer", evaluation_strategy="epoch")
```

```python
# Now finally create a "Trainer" Object with the model, training arguments, training and test datasets and evaluation function
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=small_train_dataset,
    eval_dataset=small_eval_dataset,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
# The code asks for wandb.ai login details. Details are here https://towardsdatascience.com/what-does-weights-biases-do-c060ce6b4b8e
#https://wandb.ai/deepfa
# This is used to experiment o tune up different hyperparameters
```
